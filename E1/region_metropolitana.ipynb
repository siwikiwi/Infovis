###  Pregunta 2: ¿Los productos con mayor cantidad de volumen tienen una ganacia mas alta en epocas especificas del año?


Crearemos nuevas columnas que contemplen los meses del año que corresponde cada fecha y asi agruparlo por estacion y una columnaa que me permita saebr la ganancia estimada, PERO Con los datos actuales no se puede calcular la ganancia neta, ya que no tenemos el costo de produccion o del producto, pero si calcular que productos de mayor volumen generan mas ingresos brutos por epoca del año.


```python

# 1. Convertir la columna a formato fecha nativo (súper rápido si ya tiene formato Año-Mes-Día)
frutas_rm['Fecha'] = pd.to_datetime(frutas_rm['Fecha'])

# 2. Extraer el mes y el número matemático MMDD (Mes * 100 + Día) de forma vectorizada
# Ej: 20 de marzo -> 320, 21 de junio -> 621, 21 de diciembre -> 1221
frutas_rm['Month'] = frutas_rm['Fecha'].dt.month
mmdd = frutas_rm['Fecha'].dt.month * 100 + frutas_rm['Fecha'].dt.day

# 3. Definir las reglas lógicas para todas las filas al mismo tiempo
condiciones = [
    (mmdd >= 1221) | (mmdd < 320),  # Verano: 21 Dic al 19 Mar
    (mmdd >= 320) & (mmdd < 621),   # Otoño: 20 Mar al 20 Jun
    (mmdd >= 621) & (mmdd <= 921),  # Invierno: 21 Jun al 21 Sep
    (mmdd >= 922) & (mmdd < 1221)   # Primavera: 22 Sep al 20 Dic
]

opciones = ['Verano', 'Otoño', 'Invierno', 'Primavera']

# 4. np.select evalúa las condiciones de golpe y asigna la estación en milisegundos
frutas_rm['Season'] = np.select(condiciones, opciones, default='Fecha inválida')


#Columna de ingreso bruto
frutas_rm['Ingreso Bruto'] = frutas_rm['Volumen'] * (frutas_rm['Precio promedio'])

frutas_rm
```

```python
frutas_rm[(frutas_rm["Producto"] == "Limón") & (frutas_rm["Season"] == "Invierno")]
```

```python
# Agrupar por estación y producto
top_season = (
    frutas_rm.groupby(['Season', 'Producto'], as_index=False)
      .agg({
          'Ingreso Bruto': 'sum',
          'Precio promedio': 'mean',
          'Volumen': 'sum'
      })
)

# Top 3 por estación según Ingreso Bruto
top3 = (
    top_season.sort_values(['Season', 'Ingreso Bruto'], ascending=[True, False])
              .groupby('Season')
              .head(10)
)

top3 = pd.DataFrame(top3)
top3
```

```python
top3[top3["Producto"] == "Limón"]
```

```python
lista = top3.groupby("Producto")["Ingreso Bruto"].sum().sort_values(ascending=False)
```

```python
lista
```

```python
cosas = lista.index[0:9]
```

```python
cosas
```

```python
top10 = top3[top3["Producto"].isin(cosas)]
```

```python
top10
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Definimos el orden de las estaciones para que sea consistente en ambos gráficos
orden_estaciones = top10["Season"].unique()

# Configuración de estilo visual
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# 2. LÍNEAS CON TEXTURA
sns.lineplot(
    data=top10, 
    x="Producto", 
    y="Ingreso Bruto", 
    hue="Season", 
    style="Season",
    hue_order=orden_estaciones,   # FORZAMOS EL MISMO ORDEN DE COLOR
    style_order=orden_estaciones, # FORZAMOS EL MISMO ORDEN DE TEXTURA
    legend=False,    
    alpha=0.5,       
    linewidth=2,
    palette="Set2"                # Asegúrate de usar la misma paleta en ambos
)

# 3. PUNTOS CON FORMAS
sns.scatterplot(
    data=top10, 
    x="Producto", 
    y="Ingreso Bruto", 
    hue="Season",      
    style="Season", 
    hue_order=orden_estaciones,   # MISMO ORDEN AQUÍ
    style_order=orden_estaciones, # MISMO ORDEN AQUÍ
    s=80,             
    palette="Set2",               # MISMA PALETA AQUÍ
    alpha=1.0
)

# Ajustes de etiquetas y ángulo
plt.title("Ingreso Bruto por Producto y Estación", fontsize=14, fontweight='bold', pad=15)
plt.xticks(rotation=45, ha='right', fontsize=9) # 'ha=right' alinea mejor las etiquetas rotadas
plt.yticks(fontsize=9)
plt.xlabel("Producto", fontsize=11)
plt.ylabel("Ingreso Bruto ($)", fontsize=11)

plt.legend(title="Estación", bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```
